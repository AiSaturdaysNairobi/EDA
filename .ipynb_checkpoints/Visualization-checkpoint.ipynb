```python
import pandas as pd

```

```python
busia_df = pd.read_csv('Data/BusiaTravel.csv')
busia_df.tail()
```

```python
busia_df.dtypes
```

```python
busia_df['Year'] = pd.to_datetime(busia_df['Year']) 
busia_df.drop(columns=['Year_text', 'Results_Status','OBJECTID'],inplace=True)
busia_df.head()
```

## Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
sns.set()
```

### Numerical Features

### Line Graphs
A plot used to track the change in value over time. This naturally makes it suitable for time series data.


```python

fig, axes = plt.subplots(nrows=2, ncols=2)
busia_df.plot(x='Year',y='Visitors_on_Holiday',ax=axes[0,0])
busia_df.plot(x='Year',y='Visitors_on_Business',ax=axes[0,1])
busia_df.plot(x='Year',y='Visitors_in_Transit',ax=axes[1,0])
busia_df.plot(x='Year',y='Other_Visitors',ax=axes[1,1])
```

### Area Chart
Similar to line plot but then the area and under the curve is filled. It is used to visualize trend rather than values.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
busia_df.plot(x='Year',y='Visitors_on_Holiday',ax=axes[0,0],kind='area')
busia_df.plot(x='Year',y='Visitors_on_Business',ax=axes[0,1],kind='area')
busia_df.plot(x='Year',y='Visitors_in_Transit',ax=axes[1,0],kind='area')
busia_df.plot(x='Year',y='Other_Visitors',ax=axes[1,1],kind='area')
```

### Stacked Area charts
This an area chart that shows how much each part contributes to the total.

```python
busia_df.plot(x='Year',y=['Visitors_in_Transit','Visitors_on_Holiday','Visitors_on_Business','Other_Visitors'],kind='area')
```

### Scatter Plot
A plot that checks for relationships between variables.


```python
busia_Visitors_df = busia_df.drop(columns=['Quater','Travel_Route'])

Arrivals = busia_Visitors_df[busia_df.Travel_Route=='Arrivals'].set_index('Year').groupby('Year').mean()
Departures = busia_Visitors_df[busia_df.Travel_Route=='Departures'].set_index('Year').groupby('Year').mean()
```

```python
Departures.head()
```

```python
Arrivals.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np


f = plt.figure()    
f, axes = plt.subplots(nrows = 2, ncols = 2, sharex=True)
norm=plt.Normalize(-22,22)
sc = axes[0][0].scatter(x=Arrivals.Visitors_in_Transit,y=Departures.Visitors_in_Transit, marker = "x", norm=norm)
axes[0][0].set_xlabel('Visitors in Transit', labelpad = 5)

axes[0][1].scatter(x=Arrivals.Visitors_on_Business,y=Departures.Visitors_on_Business, marker = 'o', norm=norm)
axes[0][1].set_xlabel('Visitors on Business', labelpad = 5)

axes[1][0].scatter(x=Arrivals.Visitors_on_Holiday,y=Departures.Visitors_on_Holiday, marker = '*', norm=norm)
axes[1][0].set_xlabel('Visitors on Holiday')

axes[1][1].scatter(x=Arrivals.Other_Visitors,y=Departures.Other_Visitors, marker = 's', norm=norm )
axes[1][1].set_xlabel('Other Visitors')





plt.show()
#plt.scatter(x=Arrivals.Visitors_in_Transit,y=Departures.Visitors_in_Transit)
```

### Categorical with Numerical

### Bar plots 
This is a plot that shows categorical data with their values.


```python
busia_sum_df =busia_df.groupby([busia_df.Year.dt.year,busia_df.Travel_Route]).sum()
busia_sum_df = busia_sum_df.reset_index(level='Travel_Route')
busia_sum_df.head()
```

```python

busia_sum_df[busia_sum_df['Travel_Route']=='Arrivals'].plot.bar(stacked=True)
busia_sum_df[busia_sum_df['Travel_Route']=='Departures'].plot.bar(stacked=True)
```


### Histograms
This is a bar plot of frequency distributions whose width is equal to class interval



```python
busia_df.hist()
```

## Categorical Features

### Pie Chart
A chart used to show proportions of categorical quantities


```python
busia_quarter_df = busia_df.groupby(['Quater']).sum()
busia_quarter_df
```

```python
busia_quarter_df.T.plot.pie(subplots=True)
```

### Mosaic chart
A chart that shows the relationship among two or more categorical variables

```python
pd.qcut(busia_df.Visitors_on_Holiday.values,3,labels=["low", "medium", "high"])
pd.qcut(busia_df.Visitors_on_Business,3,labels=["low", "medium", "high"])

pd.qcut(busia_df.Visitors_in_Transit.values,3,labels=["low", "medium", "high"])
pd.qcut(busia_df.Other_Visitors,3, labels=["low", "medium", "high"])
```

```python
from statsmodels.graphics.mosaicplot import mosaic
mosaic(busia_df, ['Travel_Route','Quater','Visitors_on_Holiday'], title='DataFrame as Source')
```
