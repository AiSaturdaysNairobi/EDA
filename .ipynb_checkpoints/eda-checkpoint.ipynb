A process to uncover underlying insights about the data.

Saves time

Helps in extracting and engineering features

Understand why your model fails/succeeds

Help understand validate or dispell assumptions

Can give better understanding about the domain one can easily ask the right questions

1. Visualizing data
2. Statistical summaries and inferences

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.plotly as py
import plotly.graph_objs as go
%matplotlib notebook
sns.set()
```

### Numerical Features

### Line Graphs
A plot used to track the change in value over time. This naturally makes it suitable for time series data.


```python

fig, axes = plt.subplots(nrows=2, ncols=2)
busia_df.plot(x='Year',y='Visitors_on_Holiday',ax=axes[0,0])
busia_df.plot(x='Year',y='Visitors_on_Business',ax=axes[0,1])
busia_df.plot(x='Year',y='Visitors_in_Transit',ax=axes[1,0])
busia_df.plot(x='Year',y='Other_Visitors',ax=axes[1,1])
```

### Area Chart
Similar to line plot but then the area and under the curve is filled. It is used to visualize trend rather than values.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
busia_df.plot(x='Year',y='Visitors_on_Holiday',ax=axes[0,0],kind='area')
busia_df.plot(x='Year',y='Visitors_on_Business',ax=axes[0,1],kind='area')
busia_df.plot(x='Year',y='Visitors_in_Transit',ax=axes[1,0],kind='area')
busia_df.plot(x='Year',y='Other_Visitors',ax=axes[1,1],kind='area')
```

### Stacked Area charts
This an area chart that shows how much each part contributes to the total.

```python
busia_df.plot(x='Year',y=['Visitors_in_Transit','Visitors_on_Holiday','Visitors_on_Business','Other_Visitors'],kind='area')
```

### Scatter Plot
A plot that checks for relationships between variables.


```python
busia_Visitors_df = busia_df[['Travel_Route','Visitors_on_Holiday','Visitors_on_Business','Visitors_in_Transit','Other_Visitors']]

Arrivals = busia_Visitors_df[busia_Visitors_df.Travel_Route=='Arrivals'].drop(columns='Travel_Route')
Departures = busia_Visitors_df[busia_Visitors_df.Travel_Route=='Departures'].drop(columns='Travel_Route')

#busia_Visitors_df.plot(x=Arrivals,y=Departures,subplots=True,kind='scatter')
for i in range(0,4):
    plt.scatter(x=Arrivals.iloc[:,0].values,y=Departures.iloc[:,0].values)
    
```

### Bar plots

```python
busia_sum_df =busia_df.groupby([busia_df.Year.dt.year,busia_df.Travel_Route]).sum()
busia_sum_df = busia_sum_df.reset_index(level='Travel_Route')
busia_sum_df.head()
```

### Pie Chart

### Mosaic chart
A chart that shows the relationship among two or more categorical variables

```python
temp = busia_df 
busia_categorical = temp
busia_categorical['Visitors_on_Holiday'] = pd.qcut(busia_df.Visitors_on_Holiday.values,3, 
                                                  labels=["low", "medium", "high"])
busia_categorical['Visitors_on_Business'] = pd.qcut(busia_df.Visitors_on_Business,3,
                                                   labels=["low", "medium", "high"])

busia_categorical['Visitors_in_Transit'] = pd.qcut(busia_df.Visitors_in_Transit.values,3,
                                                  labels=["low", "medium", "high"])
busia_categorical['Other_Visitors'] = pd.qcut(busia_df.Other_Visitors,3, 
                                             labels=["low", "medium", "high"])
```

```python
from statsmodels.graphics.mosaicplot import mosaic
mosaic(busia_categorical, ['Travel_Route','Quater','Visitors_on_Holiday'], title='DataFrame as Source')
```

### Dataset distribution
A data distribution is a function or a listing which shows all the possible values (or intervals) of the data. It also (and this is important) tells you how often each value occurs. 

Dataset distributions are essential because the machine learning models make assumptions about the data passed to them and this can lead to the wrong results if data is passed to an unsuitable algorithm

#### Normal Distribution
Data distribution where values cluster around the centre:
1. Symmetrical
2. Mean, median and mode are equal
3. Total area under curve is 100%

<img src="Images/normal formula.png">

Most data follow this distribution

<img src="Images/Normal.png">

#### Log Normal Distribution
A lognormal (log-normal or Galton) distribution is a probability distribution with a normally distributed logarithm. A random variable is lognormally distributed if its logarithm is normally distributed.

1. Skewed distributions with low mean values
2. large variance
3.  Values must be positive as log(x) exists only for positive values of x. 

<img src="Images/LogNormal.png">

#### Skewed Distributions

Skewness : Symmetry of data
* Where the mean is not equal to mode and median

<img src="Images/skewness.jpg">

#### Poisson Distribution
The Poisson distribution is the discrete probability distribution of the number of events occurring in a given time period, given the average number of times the event occurs over that time period. 

In addition to its use for staffing and scheduling, the Poisson distribution also has applications in biology (especially mutation detection), finance, disaster readiness, and any other situation in which events are time-independent. 

<img src="Images/Poisson.png">

#### Bernoulli Distribution

#### Binomial Distribution
A binomial distribution can be thought of as simply the probability of a SUCCESS or FAILURE outcome in an experiment or survey that is repeated multiple times. The binomial is a type of distribution that has two possible outcomes (the prefix â€œbiâ€ means two, or twice). For example, a coin toss has only two possible outcomes: heads or tails and taking a test could have two possible outcomes: pass or fail.
